```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# 데이터셋 로드
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris.data])

# 데이터셋 정규화
scaler = StandardScaler()
scaler.fit(df)
df_scaled = scaler.transform(df)

# 데이터 프레임으로 자료형 변환 및 target class 정보 추가
df_scaled = pd.DataFrame(df_scaled, columns= ['sepal length', 'sepal width', 'petal length', 'petal width'])
df_scaled['target'] = iris.target

from sklearn.decomposition import PCA

# 2차원으로 차원 축소, target 정보는 제외
pca = PCA(n_components = 2)
pca.fit(df_scaled.iloc[:,:-1])

# pca transform 후 데이터프레임으로 자료형 변경
df_pca = pca.transform(df_scaled.iloc[:,:-1])
df_pca = pd.DataFrame(df_pca, columns = ['component 0', 'component 1'])

# PCA 주성분 설명력 출력

print(pca.explained_variance_ratio_)
# [0.72962445, 0.22850762]
```

```python
import matplotlib.pyplot as plt

# class target 정보 불러오기
df_pca['target'] = df_scaled['target']

# target 별 분리
df_pca_0 = df_pca[df_pca['target'] == 0]
df_pca_1 = df_pca[df_pca['target'] == 1]
df_pca_2 = df_pca[df_pca['target'] == 2]

# target 별 시각화
plt.scatter(df_pca_0['component 0'], df_pca_0['component 1'], color = 'r', label = 'setosa')
plt.scatter(df_pca_1['component 0'], df_pca_1['component 1'], color = 'g', label = 'versicolor')
plt.scatter(df_pca_2['component 0'], df_pca_2['component 1'], color = 'b', label = 'virginica')

plt.xlabel('component 0')
plt.ylabel('component 1')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, FactorAnalysis
from sklearn.preprocessing import StandardScaler

data = load_iris()
X = StandardScaler().fit_transform(data["data"])
feature_names = data["feature_names"]

ax = plt.axes()

im = ax.imshow(np.corrcoef(X.T), cmap="RdBu_r", vmin=-1, vmax=1)

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(list(feature_names), rotation=90)
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(list(feature_names))

plt.colorbar(im).ax.set_ylabel("$r$", rotation=0)
ax.set_title("Iris feature correlation matrix")
plt.tight_layout()

```

```python
n_comps = 2

methods = [
    ("PCA", PCA()),
    ("Unrotated FA", FactorAnalysis()),
    ("Varimax FA", FactorAnalysis(rotation="varimax")),
]

for ax, (method, fa) in zip(axes, methods):
    fa.set_params(n_components=n_comps)
    fa.fit(X)

    components = fa.components_.T
    print("\n\n %s :\n" % method)
    print(components)

    data_c=X@components

    ## 시각화
    df_FA=pd.DataFrame(np.column_stack([data_c[:,0:2],iris.target ]), columns =["X1","X2","class"])

    plt.scatter(df_FA.loc[df_FA["class"] == 0]["X1"],df_FA.loc[df_FA["class"] == 0]["X2"], color = 'r', label="setosa")
    plt.scatter(df_FA.loc[df_FA["class"] == 1]["X1"],df_FA.loc[df_FA["class"] == 1]["X2"], color = 'g', label="versicolor")
    plt.scatter(df_FA.loc[df_FA["class"] == 2]["X1"],df_FA.loc[df_FA["class"] == 2]["X2"], color = 'b', label ="virginica")
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_iris
from sklearn.decomposition import FastICA, PCA

data = load_iris()
X = StandardScaler().fit_transform(data["data"])
feature_names = data["feature_names"]

ica = FastICA(n_components=2)
S_ = ica.fit_transform(X)  # 신호 재구성

components = ica.components_.T
print(components)

data_c=X@components

## 시각화
df_ICA=pd.DataFrame(np.column_stack([data_c[:,0:2],iris.target ]), columns =["X1","X2","class"])

plt.scatter(df_ICA.loc[df_ICA["class"] == 0]["X1"],df_ICA.loc[df_ICA["class"] == 0]["X2"], color = 'r', label="setosa")
plt.scatter(df_ICA.loc[df_ICA["class"] == 1]["X1"],df_ICA.loc[df_ICA["class"] == 1]["X2"], color = 'g', label="versicolor")
plt.scatter(df_ICA.loc[df_ICA["class"] == 2]["X1"],df_ICA.loc[df_ICA["class"] == 2]["X2"], color = 'b', label ="virginica")
plt.legend()
plt.show()




```

```python
import matplotlib.pyplot as plt

from sklearn import datasets, manifold

sr_points, sr_color = datasets.make_swiss_roll(n_samples=1500, random_state=0)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
fig.add_axes(ax)
ax.scatter(
    sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=50, alpha=0.8
)
ax.set_title("Swiss Roll in Ambient Space")
ax.view_init(azim=-66, elev=12)
_ = ax.text2D(0.8, 0.05, s="n_samples=1500", transform=ax.transAxes)
```

```python
sr_lle, sr_err = manifold.locally_linear_embedding(
    sr_points, n_neighbors=12, n_components=2
)

sr_tsne = manifold.TSNE(n_components=2, perplexity=40, random_state=0).fit_transform(
    sr_points
)


fig, axs = plt.subplots(figsize=(8, 8), nrows=2)
axs[0].scatter(sr_lle[:, 0], sr_lle[:, 1], c=sr_color)
axs[0].set_title("LLE Embedding of Swiss Roll")
axs[1].scatter(sr_tsne[:, 0], sr_tsne[:, 1], c=sr_color)
_ = axs[1].set_title("t-SNE Embedding of Swiss Roll")
```
