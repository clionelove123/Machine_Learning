```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

##데이터
iris = load_iris()
```

```python
## 차원축소
lda = LinearDiscriminantAnalysis(n_components=2)
X_reduced = lda.fit_transform(iris.data, iris.target)
```

```python
## 시각화
df_init=pd.DataFrame(np.column_stack([iris.data[:,1:5] ]), columns =["X1","X2","class"])
df=pd.DataFrame(np.column_stack([X_reduced ,iris.target]), columns =["X1","X2","class"])

plt.scatter(df_init.loc[df["class"] == 0]["X1"],df_init.loc[df["class"] == 0]["X2"], color = 'r', label="setosa")
plt.scatter(df_init.loc[df["class"] == 1]["X1"],df_init.loc[df["class"] == 1]["X2"], color = 'g', label="versicolor")
plt.scatter(df_init.loc[df["class"] == 2]["X1"],df_init.loc[df["class"] == 2]["X2"], color = 'b', label ="virginica")
plt.legend()
plt.show()

plt.scatter(df.loc[df["class"] == 0]["X1"],df.loc[df["class"] == 0]["X2"], color = 'r', label="setosa")
plt.scatter(df.loc[df["class"] == 1]["X1"],df.loc[df["class"] == 1]["X2"], color = 'g', label="versicolor")
plt.scatter(df.loc[df["class"] == 2]["X1"],df.loc[df["class"] == 2]["X2"], color = 'b', label ="virginica")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# 데이터셋 로드
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris.data])

# 데이터셋 정규화
scaler = StandardScaler()
scaler.fit(df)
df_scaled = scaler.transform(df)

# 데이터 프레임으로 자료형 변환 및 target class 정보 추가
df_scaled = pd.DataFrame(df_scaled, columns= ['sepal length', 'sepal width', 'petal length', 'petal width'])
df_scaled['target'] = iris.target

from sklearn.decomposition import PCA

# 2차원으로 차원 축소, target 정보는 제외
pca = PCA(n_components = 2)
pca.fit(df_scaled.iloc[:,:-1])

# pca transform 후 데이터프레임으로 자료형 변경
df_pca = pca.transform(df_scaled.iloc[:,:-1])
df_pca = pd.DataFrame(df_pca, columns = ['component 0', 'component 1'])

# PCA 주성분 설명력 출력

print(pca.explained_variance_ratio_)
# [0.72962445, 0.22850762]
```

```python
import matplotlib.pyplot as plt

# class target 정보 불러오기
df_pca['target'] = df_scaled['target']

# target 별 분리
df_pca_0 = df_pca[df_pca['target'] == 0]
df_pca_1 = df_pca[df_pca['target'] == 1]
df_pca_2 = df_pca[df_pca['target'] == 2]

# target 별 시각화
plt.scatter(df_pca_0['component 0'], df_pca_0['component 1'], color = 'r', label = 'setosa')
plt.scatter(df_pca_1['component 0'], df_pca_1['component 1'], color = 'g', label = 'versicolor')
plt.scatter(df_pca_2['component 0'], df_pca_2['component 1'], color = 'b', label = 'virginica')

plt.xlabel('component 0')
plt.ylabel('component 1')
plt.legend()
plt.show()
```
